Let's build the graph from a small code snippet using the **python_graphs** library:

```python
from python_graphs import control_flow, control_flow_graphviz
import os
from IPython.display import display, Image
```

```python
def plot_control_flow_graph(code_snippet, path):
    graph = control_flow.get_control_flow_graph(code_snippet)
    control_flow_graphviz.render(graph, include_src=code_snippet, path=path)

def create_graph(filename):
    file = open(f'{filename}.py', 'r')
    code = file.read()
    
    os.makedirs('output', exist_ok=True)
    path = f'output/{filename}_cfg.png'
    plot_control_flow_graph(code, path)
    return path
```

```python
display(Image(create_graph("code_snippet")))
```

The resulting graph has two kinds of edges: solid and dotted.

- **solid** edges connect two vertices corresponding to correct states
- **dotted** edges connect the "correct vertex" with the "incorrect vertex" (i.e. an exception occurs between them)

So we get a graph of states, paths of program execution. 

There are also two types of vertices - **terminal** (more precisely, entry and exit points) and **intermediate**. The figure shows the difference in the thickness of the borders. 

Note also that at least one solid and at least one dotted edge comes out of each intermediate node.
